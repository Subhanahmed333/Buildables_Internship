```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
#Basic
print(df.shape)
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
# Distribution of Math Scores
plt.figure(figsize=(8,5))
sns.histplot(df["math score"], kde=True, bins=20, color="skyblue")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()
```

```python
#Average Score by Gender
average_score_gender = df.groupby("gender")[["math score","reading score","writing score"]].mean()
average_score_gender
```

```python
# Visualization: average subject scores by gender
average_score_gender.plot(kind="bar",figsize=(8,6))
plt.title("Average Scores by Gender")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.show()
```

```python
# Boxplot: Scores vs Parental Education
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x="parental level of education",y="math score")
plt.title("Math Scores by Parental Education")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df[["math score","reading score","writing score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Scores")
plt.show()
```
